```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mtpl
import numpy as np
```

```python
def f(x): ## исходная ф-ция, которую нужно максимально приблизить многочленом
    return np.sin(x / 5.) * np.exp(x / 10.) + 5. * np.exp(-x / 2.)
x = np.arange(0, 16, 0.1)

M1 = np.array([[1.,1.], [1.,15.]]) # левая сторона матрицы
V1 = np.array([3.25221686527, 0.635221419579]) # правая сторона - вектор (свободеный член)

s1 = np.linalg.solve(M1,V1)

def f1(x):
    return s1[0]*1+s1[1]*x

M2 = np.array([[1.,1.,1.], [1.,8.,64.],[1.,15.,225.]]) # левая сторона матрицы
V2 = np.array([3.25221686527, 2.31617015905,0.635221419579]) # правая сторона - вектор (свободеный член)
s2 = np.linalg.solve(M2, V2)

def f2(x):
    return s2[0]*1+s2[1]*x+s2[2]*x**2


M3 = np.array([[1.,1.,1.,1.], [1.,4.,16.,64.,],[1.,10.,100.,1000.], [1., 15., 225.,3375.]]) # левая сторона матрицы
V3 = np.array([3.25221686527, 1.74684594959, 2.505416407, 0.635221419579]) # правая сторона - вектор (свободеный член)
s3 = np.linalg.solve(M3, V3)

def f3(x):
    return s3[0]*1+s3[1]*x+s3[2]*x**2+s3[3]*x**3


plt.plot(x, f(x), x, f1(x), x, f2(x), x, f3(x))
plt.grid()
plt.xlabel('input')
plt.ylabel ('output')
plt.legend(loc = 0);
plt.show()
```
