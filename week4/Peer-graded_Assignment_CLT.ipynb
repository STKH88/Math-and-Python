```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Expon RV generation + its hist and expo pdf

```python
globalsample = sts.expon.rvs(size=1000) 
x = np.linspace(0,7,100)
pdf = sts.expon.pdf(x)
plt.plot(x, pdf, label='theoretical PDF',  color = "r")
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.hist(globalsample, bins=10, normed=True, label='exponential distr', color = 'g')
plt.legend(loc='best')
plt.show()
```

## EX, DX, EX_n, DX_n generation + visuals

```python
def exp_var_sample_mean_and_plot (size):
    """Calculates the mean of the n sized exponential distributions
        and also tis function plots the distributions of sample means 
        of n-sized sample from exponential distribution, where:
        mu - is the mean of the 'parent'/whole distribution;
        sighat - is the sample standard deviation, computed by the sighat = sample.std/sqrt(size);
        size - size of the desired sample from the global distribution;
        xp - is a vector of values required to map the normal distribution curve into the hist;
        """
    infkeeper = []
    import scipy.stats as sts
    import matplotlib.pyplot as plt
    import numpy as np
    mu = globalsample.mean()
    sighat = globalsample.std()/size**0.5
    for i in range(1000):
        samplespace = sts.expon.rvs(size = size)
        samplemean = samplespace.mean()
        infkeeper.append(samplemean)
    return infkeeper, mu, sighat

def plot(dset, mu, sighat):
        plt.grid(True)
        plt.ylabel('$F(x)$')
        plt.xlabel('$x$')
        xp = np.linspace (0,3,1000)
        y = sts.norm.pdf(xp, loc = mu, scale = sighat)
        plt.legend()
        return plt.hist(dset, normed = True,color = 'g'), plt.plot(xp,y)
```

N=2

```python
dset, mu, sighat = exp_var_sample_mean_and_plot(2)
plot(dset,mu,sighat)
```
